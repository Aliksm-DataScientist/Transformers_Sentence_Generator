```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
# model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
Sentence = 'Kaizen is a greatest technique'
numberic_ids = tokenizer.encode(Sentence, return_tensors='pt')
```

```python
result = model.generate(numberic_ids, max_length=50, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
print('result', result)
```

```python
tokenizer.decode(result[0], skip_special_tokens=True)
```
